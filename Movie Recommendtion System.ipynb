```python
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("C:/Users/Admin/Downloads/movies.csv")
```

```python
movies
```

```python
selected_features = ["genres", "keywords","tagline","cast","director"]
```

```python
for i in selected_features:
    movies[i]  = movies[i].fillna('')
```

```python
movie_name = input("movie:")
```

```python
list_of_all_titles = movies["title"].tolist()
```

```python
list_of_all_titles
```

```python
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
find_close_match
```

```python
close_match = find_close_match[0]
```

```python
close_match
```

```python
index_of_movie = movies[movies.title == close_match]['index'].values[0]
```

```python
index_of_movie
```

```python
combined_features = movies['genres']+' '+movies['keywords']+' '+movies['tagline']+' '+movies['cast']+' '+movies['director']
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features) 
```

```python
similarity  = cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
```

```python
similarity_score
```

```python
sorted_similarity_score = sorted(similarity_score,key = lambda X:X[1],reverse = True )
```

```python
sorted_similarity_score
```

```python
i = 1 
for movie in sorted_similarity_score:
    index = movie[0]
    title_from_index = movies[movies.index == index]['title'].values[0]
    if(i <30):
        print(i,'.',title_from_index)
        i+=1
    
    
```
